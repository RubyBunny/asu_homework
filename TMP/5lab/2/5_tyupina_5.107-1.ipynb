```python
import timeit
import random

import numpy
import matplotlib.pyplot as plt

from Sort import quick_sort
```

```python
def generate_list(n: int) -> list:
    return [random.randint(0, n) for i in range(n)]
```

```python
def speed_tester() -> list:
    speed = []
    for i in range(1_000, 110_000, 10_000):
        arr = generate_list(i)
        speed.append(timeit.Timer(lambda: quick_sort(arr)).repeat(10, 1))
    
    return speed
```

```python
speed = speed_tester()
speed
```

```python
np_arr = numpy.array(speed)
np_arr
```

```python
np_average = numpy.mean(np_arr, axis=1)
np_average
```

```python
np_std = numpy.std(np_arr, axis=1)
np_std
```

```python
ax = plt.subplot()

x_line = range(1_000, 110_000, 10_000)

ax.plot(x_line, np_average, label="    ")
ax.plot(x_line, np_std, label="   ")

ax.set_xlabel("-  ")
ax.set_ylabel("  (s)")
ax.legend()
ax.grid()
```
