```python
START = 1000
END = 11000
STEP = 1000
```

```python
def find_speed_test(start: int, end: int, step: int) -> list[float]:
    speed_measurements = []
    S = 0
    for list_size in range(start, end+step, step):
        list_random_elements = [FuzzyNumber(randint(0, 100), randint(0, 100), randint(0, 100))for _ in range(list_size)]
        contain_list = ListContainer(list_random_elements)
        ID = contain_list[randint(0, len(contain_list)-1)][0]
        speed_measurements.append(timeit.Timer(lambda: contain_list.find(ID)).repeat(100,1))
        S += 1
    print(S)

    return speed_measurements
```

```python
find_array = find_speed_test(START, END, STEP)
print(find_array)
find_array = numpy.mean(find_array, axis=1) 
print(find_array)
```

```python
ax = plt.subplot()

x_line = range(START, END+STEP, STEP)
y_line = find_array

z = numpy.polyfit(x_line, y_line, 2)
p = numpy.poly1d(z)

plt.scatter(x_line, p(x_line))
plt.plot(x_line, p(x_line), label="Время выполнения поиска") 

plt.xlabel("Кол-во элементов")
plt.ylabel("Время")

ax.legend()
ax.grid()
```

```python
def insert_speed_test(start: int, end: int, step: int) -> list[float]:
    speed_measurements = []
    S = 0
    for list_size in range(start, end+step, step):
        list_random_elements = [
            FuzzyNumber(randint(0, 100), randint(0, 100), randint(0, 100))
            for _ in range(list_size)
        ]
        contain_list = ListContainer(list_random_elements)
        element = (0, 0, 0)
        speed_measurements.append(timeit.Timer(lambda: contain_list.insert(element)).repeat(15,1))
        S += 1
    print(S)
    return speed_measurements
```

```python
insert_array = insert_speed_test(START, END, STEP)
print(insert_array)
insert_array = numpy.mean(insert_array, axis=1) 
print(insert_array)
```

```python
ax = plt.subplot()

x_line = range(START, END+STEP, STEP)
y_line = insert_array

plt.scatter(x_line, y_line)
plt.plot(x_line, y_line, label="Время выполнения вставки") 

plt.xlabel("Кол-во элементов")
plt.ylabel("Время")

ax.legend()
ax.grid()
```

```python
def contain_speed_test(start: int, end: int, step: int) -> list[float]:
    speed_measurements = []
    S = 0
    for list_size in range(start, end+step, step):
        list_random_elements = [FuzzyNumber(randint(0, 100), randint(0, 100), randint(0, 100))for _ in range(list_size)]
        contain_list = ListContainer(list_random_elements)
        ID = contain_list[randint(0, len(contain_list)-1)][0]
        speed_measurements.append(timeit.Timer(lambda: contain_list.__contains__(ID)).repeat(100,1))
        S += 1
    print(S)
    return speed_measurements
```

```python
contain_array = contain_speed_test(START, END, STEP)
print(contain_array)
contain_array = numpy.mean(contain_array, axis=1) 
print(contain_array)
```

```python
ax = plt.subplot()

x_line = range(START, END+STEP, STEP)
y_line = contain_array

z = numpy.polyfit(x_line, y_line, 2)
p = numpy.poly1d(z)
plt.scatter(x_line, p(x_line))
plt.plot(x_line, p(x_line), label="Время выполнения проверки на наличие") 

plt.xlabel("Кол-во элементов")
plt.ylabel("Время")

ax.legend()
ax.grid()
```

```python
def delete2_speed_test(start: int, end: int, step: int) -> list[float]:
    speed_measurements = []
    S = 0
    for list_size in range(start, end+step, step):
        list_random_elements = [FuzzyNumber(randint(0, 100), randint(0, 100), randint(0, 100))for _ in range(list_size)]
        contain_list = ListContainer(list_random_elements)
        index = randint(0, len(contain_list)-1)
        speed_measurements.append(timeit.Timer(lambda: contain_list.__delitem2__(index)).repeat(100,1))
        S += 1
    print(S)
    return speed_measurements
```

```python
delete2_array = delete2_speed_test(START, END, STEP)
print(delete2_array)
delete2_array = numpy.mean(delete2_array, axis=1)
print(delete2_array)
```

```python
ax = plt.subplot()

x_line = range(START, END+STEP, STEP)
y_line = delete2_array

z = numpy.polyfit(x_line, y_line, 2)
p = numpy.poly1d(z)

plt.scatter(x_line, p(x_line))
plt.plot(x_line, p(x_line), label="Время выполнения удаления2") 

plt.xlabel("Кол-во элементов")
plt.ylabel("Время")

ax.legend()
ax.grid()
```
